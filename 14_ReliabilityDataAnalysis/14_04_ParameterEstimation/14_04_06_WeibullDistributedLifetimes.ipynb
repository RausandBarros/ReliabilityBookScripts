## Weibull Distributed Lifetimes

```r
# Package loading
library(WeibullR)
```

```r
# Data
failtime<-c(31.7, 39.2, 57.5, 65.8, 70.0, 101.7, 109.2, 130.0)
censored<-c(65.0, 75.0, 75.2, 87.5, 88.3, 94.2, 105.8, 110.0)
```

```r
# Data analysis
data<-wblr.conf(wblr.fit(wblr(failtime, censored)), lwd=1)
plot(data)
```
