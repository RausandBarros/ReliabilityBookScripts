## Empirical survivor function

```r
library(survival)
```

```r
survtime<-read.table("14_03_dataset.txt", header=FALSE, dec=".")
```

```r
# Prepare the data and calculate required values
data<-Surv(survtime[ ,'V1'])
survfunct<-survfit(Surv(survtime[ ,'V1'])~1, conf.type="none")
plot(survfunct, xlab="Time t", ylab="Survival probability")
```

**A 95% pointwise confidence interval:**

```r
survfunct<-survfit(Surv(survtime[ ,'V1'])~1, conf.type="plain")
plot(survfunct, xlab="Time t", ylab="Survival probability")
```
