## Binomial Model

```r
# Set the range (i.e., [0,1]) and the number of values to calculate
x<-seq(0, 1, length=100)
# Specify the parameters r and s
r<-2.36
s<-9.44
# Calculate the beta density for each x
y<-dbeta(x, r, s, log=FALSE)
plot(x, y, type="l", xlab=expression(theta), ylab=expression(pi(theta)))
```
